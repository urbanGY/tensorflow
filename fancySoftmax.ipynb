```python
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)
```

```python
data = np.loadtxt('wine.csv', delimiter=',', dtype=np.float32)
x_data = data[:,0:-1]
y_data = data[:,[-1]]
```

```python
nb_classes = 10
x = tf.placeholder(tf.float32, shape = [None, 11])
y = tf.placeholder(tf.int32, shape = [None, 1])
```

```python
y_one_hot = tf.one_hot(y, nb_classes)#y 가 integer 형태여야 하나봐
```

```python
y_one_hot = tf.reshape(y_one_hot, [-1, nb_classes])#ex) [[0100000000], [1000000000] ...]
```

```python
w = tf.Variable(tf.random_normal([11,nb_classes]), name='weight')
b = tf.Variable(tf.random_normal([nb_classes]), name='bias')
```

```python
logits = tf.matmul(x,w) + b #hypothesis라는 변수 안에 softmax랑 섞인 기존의 형태에서 wx+b라는 식만 logits라는 이름으로 분리
```

```python
hypothesis = tf.nn.softmax(logits)
```

```python
cost_i = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y_one_hot)
```

```python
cost = tf.reduce_mean(cost_i)
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

```python
prediction = tf.argmax(hypothesis, 1)
```

```python
correct_prediction = tf.equal(prediction, tf.argmax(y_one_hot, 1))
```

```python
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(20001):
    sess.run(optimizer, feed_dict={x:x_data, y:y_data})
    if step%1000 == 0:
        loss, acc = sess.run([cost, accuracy], feed_dict = {x:x_data, y:y_data})
        print(step, "loss : ",loss,", accuracy : ",acc)
```

```python
pred = sess.run(prediction, feed_dict={x:x_data})
for p, y in zip(pred, y_data.flatten()):#flatten은 [[1], [2] ... ] -> [1, 2, ...]이렇게 바꿔줌
    print("prediction : ",p,"y : ",int(y), "result : ",p == int(y))
```
